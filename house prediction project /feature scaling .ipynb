```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
df = train.copy()
```

```python
df
```

```python
housing = df.drop('median_house_value', axis = 1)
housing_value = df['median_house_value'].copy()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy="median")
```

```python
housing_num = housing.select_dtypes(include = [np.number])
```

```python
imputer.fit(housing_num)
```

```python
X = imputer.transform(housing_num)
```

```python
X
```

```python
housing = pd.DataFrame(X , columns = housing_num.columns, index=housing_num.index)
```

```python
housing['ocean_proximity'] = df['ocean_proximity']
```

```python
housing = housing[['ocean_proximity']]
```

```python
housing
```

```python
set(housing['ocean_proximity'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ordinal_encoder = OneHotEncoder()
```

```python
ordinal_encoder.categories_
```
