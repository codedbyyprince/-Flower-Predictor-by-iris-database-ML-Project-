```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
df = train.copy()
```

```python
df.plot(kind="scatter", y = "longitude", x = "latitude", grid = True)
```

```python
df.plot(kind="scatter", y = "longitude", x = "latitude", grid = True , alpha = 0.2)
```

```python
df.plot(kind="scatter", y = "longitude", x = "latitude", grid = True , alpha = 0.7, cmap="jet" , c="median_house_value")
```
